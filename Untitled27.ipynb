```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Define the sender and recipient email addresses
sender_email = "lakkireddy"
recipient_email = "recipient_email@example.com"

# Create a message object
msg = MIMEMultipart()
msg['From'] = sender_email
msg['To'] = recipient_email
msg['Subject'] = "Anomaly detected"

# Add a message to the email body
body = "An anomaly has been detected. Please take action."
msg.attach(MIMEText(body, 'plain'))

# Create an SMTP server instance and send the email
with smtplib.SMTP('smtp.gmail.com', 587) as smtp:
    smtp.ehlo()
    smtp.starttls()
    smtp.login(sender_email, 'your_password')
    smtp.sendmail(sender_email, recipient_email, msg.as_string())
    print("Email sent successfully")
```
